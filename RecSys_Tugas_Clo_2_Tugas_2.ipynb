```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine
from scipy.stats import pearsonr
```

```python
# Load the dataset
df = pd.read_excel('https://github.com/nanahsn/Recsys-Clo2/blob/main/Dataset%20rating%20tugas%202%20(1).xlsx?raw=true')
# Fill NaN values with 0
df.fillna(0, inplace=True)
df
```

```python
# User-based Similarity
# Menghitung kemiripan antara dua pengguna berdasarkan metrik yang berbeda
def user_similarity(user1, user2, metric):
    if metric == 'pearson':
        return pearsonr(user1, user2)[0]
    elif metric == 'cosine':
        return 1 - cosine(user1, user2)
    elif metric == 'adjusted_cosine':
        user1_avg = np.mean(user1)
        user2_avg = np.mean(user2)
        user1_adj = user1 - user1_avg
        user2_adj = user2 - user2_avg
        return 1 - cosine(user1_adj, user2_adj)
```

```python
# Item-based Similarity
# Menghitung kemiripan antara dua item berdasarkan metrik yang berbeda
def item_similarity(item1, item2, metric):
    if metric == 'pearson':
        return pearsonr(item1, item2)[0]
    elif metric == 'cosine':
        return 1 - cosine(item1, item2)
    elif metric == 'adjusted_cosine':
        non_zero_indices = np.logical_and(item1 != 0, item2 != 0)
        if np.sum(non_zero_indices) == 0:
            return 0
        item1_non_zero = item1[non_zero_indices]
        item2_non_zero = item2[non_zero_indices]
        return 1 - cosine(item1_non_zero, item2_non_zero)
```

```python
# Calculate user-based similarity matrix
user_sim_matrix = pd.DataFrame(index=df.columns[1:], columns=df.columns[1:])
for user1 in df.columns[1:]:
    for user2 in df.columns[1:]:
        if user1 == user2:
            user_sim_matrix.loc[user1, user2] = 1.0
        else:
            similarities = []
            for metric in ['pearson', 'cosine', 'adjusted_cosine']:
                similarity = user_similarity(df[user1], df[user2], metric)
                similarities.append(similarity)
            user_sim_matrix.loc[user1, user2] = similarities

# Display user-based similarity matrix
print("User-Based Similarity Matrix:")
print(user_sim_matrix)
```

```python
# Calculate item-based similarity matrix
item_sim_matrix = pd.DataFrame(index=df['Unnamed: 0'], columns=df['Unnamed: 0'])
for item1 in df['Unnamed: 0']:
    for item2 in df['Unnamed: 0']:
        if item1 == item2:
            item_sim_matrix.loc[item1, item2] = 1.0
        else:
            similarities = []
            for metric in ['pearson', 'cosine', 'adjusted_cosine']:
                # Extract ratings for both items
                item1_ratings = df.loc[df['Unnamed: 0'] == item1].values[0][1:]
                item2_ratings = df.loc[df['Unnamed: 0'] == item2].values[0][1:]

                # Check if both items have ratings from at least 2 users
                common_users = np.logical_and(item1_ratings > 0, item2_ratings > 0)
                if np.sum(common_users) >= 2:
                    similarity = item_similarity(item1_ratings[common_users], item2_ratings[common_users], metric)
                    similarities.append(similarity)
                else:
                    similarities.append(0)

            item_sim_matrix.loc[item1, item2] = similarities

# Display item-based similarity matrix
print("\nItem-Based Similarity Matrix:")
print(item_sim_matrix)
```

```python
# Calculate user-based similarity matrix (Pearson Correlation)
user_sim_matrix = pd.DataFrame(index=df.columns[1:], columns=df.columns[1:])
for user1 in df.columns[1:]:
    for user2 in df.columns[1:]:
        if user1 == user2:
            user_sim_matrix.loc[user1, user2] = 1.0
        else:
            similarity = pearsonr(df[user1], df[user2])[0]
            user_sim_matrix.loc[user1, user2] = similarity

# Set the similarity threshold
threshold = 0.5

# Function to predict a user's rating for an item
def predict_rating(user, item):
    numerator = 0
    denominator = 0
    for other_user in df.columns[1:]:
        if other_user != user:
            similarity = user_sim_matrix.loc[user, other_user]
            if similarity > threshold:
                other_user_rating = df.at[df[df['Unnamed: 0'] == item].index[0], other_user]
                if other_user_rating != 0:
                    numerator += similarity * other_user_rating
                    denominator += similarity
    if denominator == 0:
        return 0  # Unable to make a prediction
    else:
        return numerator / denominator

# Example Predict User rating for Item1
user = 'User1'
item = 'Item1'
predicted_rating = predict_rating(user, item)
print(f"Predicted rating for {item} by {user}: {predicted_rating}")
```
